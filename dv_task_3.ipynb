```python
!pip install joypy
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.graphics.mosaicplot import mosaic

# --- Step 2: Load Dataset ---
# Upload file manually in Colab before running
from google.colab import files
uploaded = files.upload()

# Replace with actual filename after upload
df = pd.read_csv("hospital data analysis (1).csv")

# --- Step 3: Set Style ---
sns.set(style="whitegrid")

# --- Step 4: Categorical vs Categorical ---

# 1. Stacked Bar Chart - Gender vs Readmission
cross_tab = pd.crosstab(df['Gender'], df['Readmission'])
cross_tab.plot(kind='bar', stacked=True, colormap="Pastel1", figsize=(7,5))
plt.title("Stacked Bar Chart - Gender vs Readmission")
plt.ylabel("Count")
plt.show()

# 2. Grouped Bar Chart - Outcome by Gender
plt.figure(figsize=(7,5))
sns.countplot(x="Outcome", hue="Gender", data=df, palette="Set2")
plt.title("Grouped Bar Chart - Outcome by Gender")
plt.show()

# 3. Segmented Bar Chart - Outcome by Readmission
segmented = pd.crosstab(df['Outcome'], df['Readmission'], normalize='index')
segmented.plot(kind='bar', stacked=True, figsize=(7,5), colormap="Accent")
plt.title("Segmented Bar Chart - Outcome vs Readmission (Proportions)")
plt.ylabel("Proportion")
plt.show()

# 4. Mosaic Plot - Gender vs Outcome
plt.figure(figsize=(8,6))
mosaic(df, ['Gender','Outcome'])
plt.title("Mosaic Plot - Gender vs Outcome")
plt.show()

```

```python
# --- Step 5: Continuous vs Continuous ---
sns.lmplot(x="Age", y="Cost", data=df, height=5, aspect=1.5, scatter_kws={'alpha':0.5})
plt.title("Scatterplot with Fit Line - Age vs Cost")
plt.show()
```

```python
# Task 3: Visualizations for Categorical vs Continuous and Continuous vs Continuous

# --- Step 1: Import Libraries ---
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# For Ridgeline Plot
!pip install joypy
from joypy import joyplot

# --- Step 2: Load Dataset ---
# Upload file manually in Colab before running
from google.colab import files
uploaded = files.upload()

# Replace with actual filename after upload
df = pd.read_csv("hospital data analysis (1).csv")

# --- Step 3: Set Style ---
sns.set(style="whitegrid")

# --- Step 4: Categorical vs Continuous Plots ---

plt.figure(figsize=(15,10))

# 1. Bar Chart - Average Cost by Gender
plt.subplot(2,3,1)
sns.barplot(x="Gender", y="Cost", data=df, estimator="mean", palette="Set2")
plt.title("Bar Chart - Average Cost by Gender")

# 2. Grouped Kernel Density Plot - Age distribution by Gender
plt.subplot(2,3,2)
sns.kdeplot(data=df, x="Age", hue="Gender", fill=True, alpha=0.5)
plt.title("Grouped KDE - Age by Gender")

# 3. Box Plot - Cost by Outcome
plt.subplot(2,3,3)
sns.boxplot(x="Outcome", y="Cost", data=df, palette="Set3")
plt.title("Box Plot - Cost by Outcome")

# 4. Violin Plot - Satisfaction by Outcome
plt.subplot(2,3,4)
sns.violinplot(x="Outcome", y="Satisfaction", data=df, palette="muted")
plt.title("Violin Plot - Satisfaction by Outcome")

# 5. Ridgeline Plot - Age distribution by Condition
plt.subplot(2,3,5)
joyplot(data=df, by="Condition", column="Age", figsize=(10,6), colormap=plt.cm.Set3)
plt.title("Ridgeline Plot - Age by Condition")
plt.xlabel("Age")

# 6. Beeswarm Plot - Cost by Readmission
plt.subplot(2,3,6)
sns.swarmplot(x="Readmission", y="Cost", data=df, palette="husl")
plt.title("Beeswarm Plot - Cost by Readmission")

plt.tight_layout()
plt.show()
```
