```python
# Task 2: Visualizations of Categorical and Continuous Data

# --- Step 1: Import Libraries ---
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# --- Step 2: Load Dataset ---
# Upload file manually in Colab before running
from google.colab import files
uploaded = files.upload()

# Replace with actual filename after upload
df = pd.read_csv("/content/hospital data analysis (1) (1).csv")

# --- Step 3: Set Style ---
sns.set(style="whitegrid")

# --- Step 4: Categorical Data ---
plt.figure(figsize=(12,5))

# Bar Chart - Count of Patients by Gender
plt.subplot(1,2,1)
sns.countplot(x="Gender", data=df, palette="Set2")
plt.title("Bar Chart - Gender Distribution")

# Pie Chart - Condition Distribution
plt.subplot(1,2,2)
df['Condition'].value_counts().plot.pie(autopct='%1.1f%%', startangle=90, cmap="Set3")
plt.ylabel("")
plt.title("Pie Chart - Condition Distribution")
plt.show()

# --- Step 5: Continuous Data ---
plt.figure(figsize=(15,10))

# Scatterplot - Age vs. Cost
plt.subplot(2,3,1)
sns.scatterplot(x="Age", y="Cost", data=df, alpha=0.6)
plt.title("Scatterplot - Age vs Cost")

# Line Plot - Average Satisfaction by Age
plt.subplot(2,3,2)
sns.lineplot(x="Age", y="Satisfaction", data=df, ci=None)
plt.title("Line Plot - Age vs Satisfaction")

# Strip Plot - Length of Stay by Gender
plt.subplot(2,3,3)
sns.stripplot(x="Gender", y="Length_of_Stay", data=df, jitter=True)
plt.title("Strip Plot - Length of Stay by Gender")

# Swarm Plot - Satisfaction by Outcome
plt.subplot(2,3,4)
sns.swarmplot(x="Outcome", y="Satisfaction", data=df)
plt.title("Swarm Plot - Satisfaction by Outcome")

# Histogram - Age distribution
plt.subplot(2,3,5)
sns.histplot(df["Age"], bins=20, kde=False, color="skyblue")
plt.title("Histogram - Age Distribution")

# Density + Rug Plot - Cost
plt.subplot(2,3,6)
sns.kdeplot(df["Cost"], shade=True, color="red")
sns.rugplot(df["Cost"], color="black")
plt.title("Density & Rug Plot - Cost")
plt.show()
```
